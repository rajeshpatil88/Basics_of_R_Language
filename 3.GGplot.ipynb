### 2. Visualizing categorical attributes using bar plots

```r
##Grouped bar plots

counts = table(drive_wheel,aspiration)
counts
```

```r
barplot(counts, 
        main="Distribution of drive-wheel in each aspiration type",
        col=rainbow(7),
        xlab="Engine Aspiration", 
        ylab = "num_Cars",
        legend.text = TRUE,
        args.legend = list(x = "topright", bty = "n", cex = 2, ncol=1),
        beside = TRUE)
```

__Differences between histograms and bar charts:__  
1. **Histograms** are used to show distributions of variables while **Bar charts** are used to compare variables.   
2. **Histograms** plot binned quantitative data while bar charts plot categorical data. 
3. **Bars** can be reordered in bar charts but not in histograms.

2. Noun -> Data 
3. Verb -> "geom_" + plot type
4. Adjectives -> Aesthetics (x,y,color, shape, size, fill, etc)

__Bar plots__
##### Start with a basic graph  in ggplot
##### If y is a numeric vector, use stat="identity"; the default is stat="bin" for categorical

 

```r
#Add title to ggplot
ggplot(cars_data,aes(x=drive_wheel)) + 
geom_bar() +
xlab("gear") + 
ylab("Cars")  +
ggtitle("gear in various cats")
```

```r
# Fill color based on number of cylinders
ggplot(cars_data,
       aes(x=drive_wheel, fill=aspiration)) +
  geom_bar(colour = "red") +
  xlab("gear") + 
  ylab("crab")  +
  ggtitle("#gear on different carb")
```

```r
names(cars_data)
```

```r
ggplot(cars_data,aes(x=drive_wheel)) +
  geom_bar(aes(fill=num_cyl)) +
        xlab("Fuel_system") + ylab("# Cars")  +
        ggtitle("#Cars in various fuel system types") +
  scale_fill_manual(values=c("blue4", "orangered2", "red",
                            "goldenrod", "darkolivegreen1", "black", 'firebrick'))
```

```r
# Make a grouped bar chart with similar graph attributes
ggplot(cars_data,aes(x=drive_wheel, fill=num_cyl)) + 
        geom_bar(position = "dodge") +
        xlab("Fuel_system") + 
        ylab("# Cars")  +
        ggtitle("#Cars in various fuel system types")
```

```r
# To adjust the width of the bars
ggplot(cars_data,aes(x=drive_wheel,fill=num_cyl)) +
        geom_bar(width = 0.5) +
        xlab("Fuel_system") + ylab("# Cars")  +
        ggtitle("#Cars in various fuel system types")
```

```r
# To add a black outline to the bars
ggplot(cars_data,aes(x=fuel_system,
                     fill=num_cyl)) + 
        geom_bar(width = 0.5, colour = "black" ) +
        xlab("Fuel_system") + ylab("# Cars")  +
        ggtitle("#Cars in various fuel system types")
```

```r
# Change back ground
#To remove grey background in plot
ggplot(cars_data,aes(x=fuel_system,
                     fill=num_cyl)) +
        geom_bar(width = 0.5,colour = "black") +
        xlab("Fuel_system") + ylab("# Cars")  +
        ggtitle("#Cars in various fuel system types")+
        theme_bw()
```

```r
# To remove grey and gridlines in the Bg
ggplot(cars_data,aes(x=fuel_system,
                     fill=num_cyl)) +
        geom_bar(width = 0.5) +
        xlab("Fuel_system") + ylab("# Cars")  +
        ggtitle("#Cars in various fuel system types")+
        theme_classic()
```

```r
# Adjust text size and angle of labels
ggplot(cars_data,aes(x=fuel_system,
                     fill=num_cyl)) +
        geom_bar(width = 0.5,colour = "black") +
        xlab("Fuel_system") + 
        ylab("# Cars")  +
        ggtitle("#Cars in various fuel system types")+
        theme_classic()+
        theme(axis.text.x=element_text(angle=45,
                                       size=12),
              text=element_text(size=14))
```

```r
# Now remove the legend position
ggplot(cars_data,aes(x=fuel_system,
                     fill=num_cyl)) +
        geom_bar(width = 0.5) +
        xlab("Fuel_system") + ylab("# Cars")  +
        ggtitle("#Cars in various fuel system types")+
        theme_classic()+
        theme(axis.text.x=element_text(angle=45,size=12),
              text=element_text(size=10), 
              legend.position = "none")
```

```r
ggplot(cars_data,aes(x= highway_mpg,y = aspiration)) + 
  geom_boxplot(width=0.5,lwd=1)
```

```r
# Scatterplot with 3 dimensions (by adding color to the basic plot)
ggplot(data=cars_data, aes(x=city_mpg,
                           y=price, 
                           colour = aspiration)) + 
        geom_point() + 
        xlab("City_mpg")+ylab("Price") + 
        ggtitle("City_mpg Vs. Car Price")
```

```r
# Scatterplot with 5 dimensions (by adding color, shape, & size to the basic plot)
ggplot(data=cars_data, aes(x=city_mpg, 
                           y=price,
                           colour = aspiration, 
                           size = hoesepower, 
                           shape = drive_wheel)) + 
        geom_point() + 
        xlab("City_mpg")+ylab("Price") + 
        ggtitle("City_mpg Vs. Car Price")
```

```r
#Jitter widens the point space for each level
ggplot(data=cars_data, aes(x=drive_wheel, y=price, colour = aspiration, size = hoesepower)) + 
        geom_jitter() + 
        xlab("City_mpg")+ylab("Price") + 
        ggtitle("drive_wheel Vs. Car Price") 
```
